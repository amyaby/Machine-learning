## Train → Fit → Predict

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Sakshi2k/Position-Salaries/refs/heads/master/Position_Salaries.csv'
dataset = pd.read_csv(url)
```

```python
dataset.info()
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
# Take all data, split into 80% train, 20% test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
def viz_linear():
    plt.scatter(X,y,color='red')
    plt.plot(X,lin_reg.predict(X),color='blue')
    plt.title('linear Regression model')
    plt.xlabel('position level')
    plt.ylabel('Salary')
    plt.show()
    return
```

```python
viz_linear()
```

the points are not aligned we will try polynomial regression and see the key differences 

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree =4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly,y)
```

```python
LinearRegression()
```

```python
def viz_polynomial():
    plt.scatter(X,y,color ='red')
    plt.plot(X,pol_reg.predict(poly_reg.fit_transform(X)),color='blue')
    plt.title('linear regression with polynomial with degree 4')
    plt.xlabel('position level')
    plt.ylabel('salary')
    plt.show()
    return
viz_polynomial()
```
