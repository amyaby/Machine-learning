```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(iris.data)
print(iris.target)
```

```python
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']

y = pd.DataFrame(iris.target)
y.columns = ['class']
```

```python
plt.scatter(x.Petal_Length, x.Petal_Width)
plt.show()
```

```python
colorL = np.array(['green', 'red', 'blue'])

plt.scatter(x.Petal_Length, x.Petal_Width, c=colorL[iris.target])
plt.show()
```

```python
inert = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    inert.append(kmeans.inertia_)

plt.plot(range(1, 11), inert)
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
plt.scatter(x.Petal_Length, x.Petal_Width, c=colorL[model.labels_])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(x.Petal_Length, x.Petal_Width, c=colorL[model.labels_])
ax1.set_title('Clusters')

ax2.scatter(x.Petal_Length, x.Petal_Width, c=colorL[iris.target])
ax2.set_title('Vraies classes')

plt.show()
```
